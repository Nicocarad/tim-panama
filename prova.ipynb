## Test PDH

```python
pdh = pd.read_parquet("alarms datasets/tx/20230101-20240101_inpas_tx_preprocess__pdh__last_event__last_event__ext1.parquet")
```

```python
pdh = pdh[['alarm_id', 'std_probable_cause_no']]
```

```python
# Trova le righe in cui 'std_probable_cause_no' Ã¨ null
rows_with_null = pdh[pdh['std_probable_cause_no'].isnull()]
rows_with_null
```

```python
mob = mob.rename(columns={'std_probable_cause_no': 'slogan'})
adsl = adsl.rename(columns={'std_probable_cause_no': 'slogan'})
sdh = sdh.rename(columns={'std_probable_cause_no': 'slogan'})
ptn = ptn.rename(columns={'std_probable_cause_no': 'slogan'})
pdh = pdh.rename(columns={'std_probable_cause_no': 'slogan'})
```

```python
total_alarms = pd.concat([mob, sdh, ptn, adsl, pdh])
```

```python
total_alarms
```

```python
cluster_filtered = real_time_clusters[["alarm_id", "cluster_id2","ne_id","domain"]]
```

```python
cluster_filtered
```

```python
cluster_merge = pd.merge(cluster_filtered, total_alarms, on='alarm_id', how='left')
```

```python
num_rows = cluster_merge['slogan'].isnull().sum()
num_rows
```

```python
# Calcola il numero di valori univoci nella colonna 'slogan'
num_unique_values = cluster_merge['slogan'].nunique()
```

```python
num_unique_values
```

```python
# Conta il numero di volte che ogni valore unico nella colonna 'slogan' appare
slogan_counts = cluster_merge['slogan'].value_counts()
```

```python
slogan_counts
```

```python
cluster_merge
```

```python
clusters = cluster_merge.groupby(['cluster_id2'])
```

```python
clusters_filtered = clusters.filter(lambda x: len(x) >= 3)
```

```python
clusters_filtered
```

```python
clusters_filtered_grouped = clusters_filtered.groupby(["cluster_id2"])
```

```python
clusters_filtered = clusters_filtered_grouped.filter(lambda x: x['ne_id'].nunique() >= 2)
```
